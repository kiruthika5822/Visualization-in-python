# Visualization

## ABOUT THE DATASET

### This dataset is downloaded from the Kaggle dataset website and used it to do some basic visualization using pandas.The following is a Worldbank dataset on education expenditures of various countries from the year 2010 to 2021.Expenditure on education, total (% of government expenditure) is calculated by dividing total government expenditure on education by the total government expenditure on all sectors and multiplying by 100.Here the values given are in billion dollars.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### To perform the operations we need to import a library, so here pandas,matplotlib,seaborn libraries are imported locally as pd,plt,sns.

```python
ds=pd.read_csv("D:\SEMESTER II\Python\datasets\World Bank Data on Education Expenditure.csv")
```

#### Use pd.read_csv function to read the dataset in to Jupyter Notebook

```python
ds
```

#### The dataset is viewed and we can see information that are contained in this dataset

```python
plt.hist(ds['2010'],color='pink')
```

#### From the histogram, Educational expenditure on more than 50% of the countries lies between 10-20 Billion Dollars. 

```python
plt.scatter(ds['2010'],ds['2021'],color='yellow')
```

#### From the year 2010 to 2021 there is a gradual increase in all country's  educationalÂ expenditures.

```python
sns.distplot(ds['2010'],color='purple',rug=True)
plt.show()
```

#### rug lines are clustered in some areas. Density of values are higher inbetween 10 to 20 and reduce at 30 again increase at 36 - 37.

```python
plt.subplot(2,2,1)
plt.title('2019')
sns.distplot(ds['2019'])

plt.subplot(2,2,2)
plt.title('2021')
sns.distplot(ds['2021'])

plt.subplot(2,2,3)
plt.title('2020')
sns.distplot(ds['2020'])
plt.tight_layout()

plt.show()
```

#### Expenditure on education, total (% of government expenditure) is calculated by dividing total government expenditure on education by the total government expenditure on all sectors and multiplying by 100. Because of covid lockdown , in 2020 whole world faced problem in running each and every sector. So expenditures by government on any sector including educational sector is less compared to previous year. In this scenario world bank expense on education is also slightly reduced in 2021 based on previous year expenditure by countries.

```python
ds['average'] = ds.iloc[:, 12:].astype(float).mean(axis=1)
```

```python
ds
```

#### Calculating average for each country's educational expense by world bank.

```python
sns.jointplot(x='average',y='2021',data=ds,kind='hex',color='crimson')
plt.show()
```

#### Comparing histograms of 2021 and average values countries with 8 to 13 billion dollars expenditure is increased in 2021 than average expenditure.

#### Comparing density of all years from 2010 to 2021.

```python
ds1=ds.loc[(ds['2020']>=14.5) & (ds['2021']>=14.5)]
ds2=ds1.sort_values('2021', ascending = False)
```

#### Filtering expenditures, greater than average and arrange it in descending order to show which countries got higher expenditure cost from world bank.

```python
x=list(ds2.iloc[:,0])
y=list(ds2.iloc[:,12])
plt.bar(x,y,color='lightgreen')
plt.title('Country_Name vs 2021')
plt.xlabel('Country_Name')
plt.ylabel('2021')
plt.xticks(color='black',rotation=90,horizontalalignment='right')
```

#### Plotting the above filtered expenditures.

```python
sns.boxplot(ds['average'],color='turquoise')
plt.title('average')
plt.show()
```
